Data preprocessing

```python
df = pd.read_csv('Market_Basket_Optimisation_Apriori.csv', header = None)

transactions = []
for i in range(0, 7501):
  transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

Training the apriori model on dataset

```python
from apyori import apriori

rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, lift = 3, min_length = 2, max_length = 2)
```
